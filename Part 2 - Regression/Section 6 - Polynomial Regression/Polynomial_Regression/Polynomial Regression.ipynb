```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Position_Salaries.csv")
df.head()
```

```python
X = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,4))
plt.scatter(X,y,c='r',marker = '*',label='Observed')
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

```python
LR = LinearRegression()
model = LR.fit(X,y)
predict = model.predict(X)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,4))
plt.scatter(X,y,c='r',marker = '*',label='Observed Data')
plt.plot(X,predict,c='g',label='Linear Model')
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_model = PolynomialFeatures(degree=4)
X_poly = poly_model.fit_transform(X)
poly_model.fit(X_poly,y)
```

```python
X_train = X_poly[0:7]
y_train = y[0:7]
X_test = X_poly[7:10]
y_test = y[7:10]
```

```python
LR_poly = LinearRegression()
model_poly = LR_poly.fit(X_train,y_train)
predict_poly = model_poly.predict(X_test)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,4))
plt.scatter(X[7:10],y_test,c='r',marker = '*',label='Observed Data')
plt.plot(X[7:10],predict_poly,c='g',label='Linear Model')
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```
